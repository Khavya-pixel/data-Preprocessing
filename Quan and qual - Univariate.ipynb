```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Min","25%","50%","75%","100%","IQR",
                               "1.5IQR","LesserRange","GreaterRange"],columns=quan_cols)
for columnName in quan_cols:
    descriptive[columnName]["Mean"]=df[columnName].mean()
    descriptive[columnName]["Median"]=df[columnName].median()
    descriptive[columnName]["Mode"]=df[columnName].mode()[0]
    descriptive[columnName]["Min"]=df[columnName].min()
    descriptive[columnName]["25%"]=df.describe()[columnName]["25%"]
    descriptive[columnName]["50%"]=df.describe()[columnName]["50%"]
    descriptive[columnName]["75%"]=df.describe()[columnName]["75%"]
    descriptive[columnName]["100%"]=df.describe()[columnName]["max"]
    descriptive[columnName]["IQR"]=descriptive[columnName]["75%"]-descriptive[columnName]["25%"]
    descriptive[columnName]["1.5IQR"]=1.5*descriptive[columnName]["IQR"]
    descriptive[columnName]["LesserRange"]=descriptive[columnName]["25%"]- descriptive[columnName]["1.5IQR"]
    descriptive[columnName]["GreaterRange"]=descriptive[columnName]["75%"]+ descriptive[columnName]["1.5IQR"]
   
```

```python
def describe(columnName,df):
    descriptive=pd.DataFrame(index=["Mean","Median","Mode","Min","25%","50%","75%","100%","IQR",
                                   "1.5IQR","LesserRange","GreaterRange",'Skewness','Kurtosis'],columns=quan_cols)
    for columnName in quan_cols:
        descriptive[columnName]["Mean"]=df[columnName].mean()
        descriptive[columnName]["Median"]=df[columnName].median()
        descriptive[columnName]["Mode"]=df[columnName].mode()[0]
        descriptive[columnName]["Min"]=df[columnName].min()
        descriptive[columnName]["25%"]=df.describe()[columnName]["25%"]
        descriptive[columnName]["50%"]=df.describe()[columnName]["50%"]
        descriptive[columnName]["75%"]=df.describe()[columnName]["75%"]
        descriptive[columnName]["100%"]=df.describe()[columnName]["max"]
        descriptive[columnName]["IQR"]=descriptive[columnName]["75%"]-descriptive[columnName]["25%"]
        descriptive[columnName]["1.5IQR"]=1.5*descriptive[columnName]["IQR"]
        descriptive[columnName]["LesserRange"]=descriptive[columnName]["25%"]- descriptive[columnName]["1.5IQR"]
        descriptive[columnName]["GreaterRange"]=descriptive[columnName]["75%"]+ descriptive[columnName]["1.5IQR"]
        descriptive[columnName]["skewness"]=df[columnName].skew()
        descriptive[columnName]["Kurtosis"]=df[columnName].kurtosis()
    return descriptive
```
